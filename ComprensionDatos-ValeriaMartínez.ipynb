Se puede observar que hay más datos atipicos en la relación que existe entre el outcome 1 en donde la persona es diabetica con la variable pedigree function que interpreta si las posibilidades de tener diabetes vienen de historia familiar. 

```python
correlation_matrix = diabetes.corr().round(2)
```

```python
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
corr=diabetes.corr()#["DiabetesPedigreeFunction"]
plt.figure(figsize=(10, 10))

sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features');
```
